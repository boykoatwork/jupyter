```python
import scipy.stats

s = np.random.normal(0, .5, 1000)
t = np.random.rand(1000)
plot(t,'bo',s,'ro')
plt.show() # для двух графиков сразу
fit = sc.stats.norm.pdf(s, np.mean(s), np.std(s))  #this is a fitting indeed
```

```python
s = np.random.normal(1, .1, 1000)
t = np.random.rand(1000)
plot(t,'bo',s,'ro')
```

Обратите внимание, что время может быть придется отсеивать по отрицательным величинам - некоторые результаты может "забрасывать" за ноль.

```python
s = np.random.weibull(1.,1000)
t = np.random.rand(1000)
plot(t,'bo',s,'ro')
```

```python
s = 2 - np.random.weibull(1.,1000)
t = np.random.rand(1000)
plot(t,'bo',s,'ro')
```

```python
s = np.random.poisson(1.,1000)
t = np.random.rand(1000)
plot(t,'bo',s,'ro')
```

```python
s = np.random.weibull(4., 1000)
t = np.random.rand(1000)
plot(t,'bo',s,'ro')
```

```python
s = np.random.weibull(5., 1000)
t = np.random.rand(1000)
plot(t,'bo',s,'ro')
```

```python
s_w = np.random.weibull(5., 1000)*1000
t = np.random.rand(1000)*1000
plot(t,'bo',s_w,'ro')
```

```python
s_w = np.random.exponential(1, 1000)
t = np.random.rand(1000)
plot(t,'bo',s_w,'ro')
```
