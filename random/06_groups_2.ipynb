```python
import random
import math
```

```python
x = []
for i in range(100):
    x.append(random.random())
```

```python
y = []
for i in range(len(x)):
    y.append(math.sqrt(x[i] * 2 / 0.001))
```

```python
plot(y)
```

```python
min(y)
```

```python
max(y)
```

```python
median(y)
```

# vim vs emacs

```python
pol = np.random.normal(0, .2, 1000)
plot(pol,'go',label='test1')
#plt.xlabel('Foo')
plt.ylabel(u'Приверженность')
plt.title(u"Отдельные люди")
plt.plot([0, 1000], [0, 0], 'k-', lw=2)
```

```python
pol_plus = np.random.normal(.5, .2, 500)
pol_minus = np.random.normal(-.5, .2, 500)
plot(pol_plus,'ro', pol_minus, 'bo',label=['test1','2'])
plt.ylabel(u'Приверженность')
plt.title(u"Поляризация: vim vs emacs")
plt.plot([0, 500], [0.5, 0.5], 'k-', lw=2)
plt.plot([0, 500], [-0.5, -0.5], 'k-', lw=2)
plt.plot([0, 500], [0, 0], 'k--', lw=2)
```

```python
pol_plus = np.random.normal(.5, .2, 500)
pol_minus = np.random.normal(-.5, .2, 500)
tea = np.random.normal(.5, .2, 500)
coffe = np.random.normal(-.5, .2, 500)
plot(pol_plus,tea,'ro', pol_minus,coffe, 'bo',label=['test1','2'])
plt.ylabel(u'Приверженность')
plt.xlabel(u'Чай против кофе')
plt.title(u"Поляризация чая против кофе и vim vs emacs")
```

```python
pol_plus = np.random.normal(.5, .2, 500)
pol_minus = np.random.normal(-.5, .2, 500)
tea = np.random.normal(.5, .2, 500)
coffe = np.random.normal(-.5, .2, 500)
plot(pol_plus,tea+coffe,'ro',  pol_minus,tea+coffe, 'bo',label="debm")
plt.ylabel(u'Редактор')
plt.xlabel(u'Чай или Кофе')
plt.title(u"Поляризация")
```
