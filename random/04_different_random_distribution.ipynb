# Сравнительное представление различных функций распределения

Функцией распределения в теории вероятностей — функция, характеризующая распределение случайной величины или случайного вектора; вероятность того, что случайная величина X примет значение, меньшее х, где х — произвольное действительное число. При соблюдении известных условий (см. ниже) полностью определяет случайную величину. 

```python
import pylab
import random

```

```python
x = []
y = []

for i in range(1000):
    x.append(random.random())
    y.append(random.random())
```

```python
pylab.plot(x,y,'or')
```

```python
import numpy
s = numpy.random.laplace(0.1, 1, 1000)
s2 = numpy.random.weibull(1.5, 1000)
```

```python
pylab.plot(x,s,'.b')
pylab.plot(x,y,'.r')
pylab.plot(x,s2,'.y')
```

```python
import numpy
s = numpy.random.weibull(1.5, 1000)
```

```python
pylab.plot(x,s,'.b')
```

```python
pylab.plot(s,y,'.b')
```

```python
s = numpy.random.weibull(0.5, 1000)
pylab.plot(s,y,'.b')

s2 = numpy.random.weibull(1., 1000)
pylab.plot(s2,y,'.r')

s3 = numpy.random.weibull(1.5, 1000)
pylab.plot(s3,y,'.g')
s4 = numpy.random.weibull(3, 1000)
pylab.plot(s4,y,'.y')
```
