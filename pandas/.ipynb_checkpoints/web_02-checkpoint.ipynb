```python
import pandas as pd
```

```python
w = pd.read_csv('data/access-log', 
                sep=' ', 
                index_col=False, 
                names=['ip', '1', '2', 'date', 'tz', 'get', 'status', 'bytes'])
```

```python
w2 = w[['ip']]
```

```python
w2
```

```python
w['bytes'] = w['bytes'].str.replace('-','0')
```

```python
w[w['status'] == 304]
```

```python
w2['bytes'] = pd.to_numeric(w['bytes'], downcast="integer")
```

```python
w2
```

```python
w2['bytes'].sum()/(1024 * 1024) # всего трафика потреблено в Мегабайтах
```

```python
w2['bytes'].mean()/1024 # килобайт на 1 запрос
```

```python
stat = w2[['ip', 'bytes']].groupby('ip').sum()
```

```python
stat['Kb'] = stat['bytes']/1024
```

```python
stat['Mb'] = stat['Kb']/1024
```

```python
stat.sort_values(by = 'bytes', ascending = False).head(40)
```

```python
stat = w2[['ip', 'bytes']].groupby('ip').count()
```

```python
stat.sort_values(by = 'bytes')
```

```python
w2
```

```python
stat1 = w2[['ip', 'bytes']].groupby('ip').sum() / 1024
```

```python
stat1.rename(columns = {'bytes':'Kb'}, inplace = True)
```

```python
stat2 = w2[['ip', 'bytes']].groupby('ip').count()
```

```python
stat2.rename(columns = {'bytes':'clicks'}, inplace = True)
```

```python
stat3 = w2[['ip', 'bytes']].groupby('ip').mean()
```

```python
stat3.rename(columns = {'bytes':'bytes_2_clicks'}, inplace = True)
```

```python
stat = pd.concat([stat1, stat2, stat3], axis = 1)
```

```python
stat
```

```python
stat.sort_values(by = 'bytes_2_clicks', ascending = False)
```
