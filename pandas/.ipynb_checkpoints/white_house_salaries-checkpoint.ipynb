```python
df['SALARY']
```

```python
clean = df['SALARY'].str.replace('$','', regex=False).str.replace(',','', regex=False)
```

```python
salary = pd.to_numeric(clean)
```

```python
salary
```

```python
salary.sum()
```

```python
salary.mean()
```

```python
df['s'] =  salary
```

```python
df
```
