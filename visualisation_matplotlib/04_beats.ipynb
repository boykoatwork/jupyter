## Простые гармонические функции

Вернемся к предыдущему занятию. Попробуем создать графики простейших гармонических функций - синусоиды и косинусоиды.

```python
from numpy import *
import matplotlib.pyplot as plt

a1 = 1.0
a2 = 1.0
w1 = 1.0
w2 = 1.0
t = linspace(0,17,num=300)
x1 = a1 * sin(w1 * t)
x2 = a2 * cos(w2 * t)
plt.plot(t,x1)
plt.plot(t,x2)
```

Проведем несколько наглядных изменений частот и амплитут синусоид:

```python
from numpy import *
import matplotlib.pyplot as plt

a1 = 1.0
a2 = 1.0
w1 = 1.0
w2 = 3.0
t = linspace(0,17,num=300)
x1 = a1 * sin(w1 * t)
x2 = a2 * cos(w2 * t)
plt.plot(t,x1)
plt.plot(t,x2)
```

```python
from numpy import *
import matplotlib.pyplot as plt

a1 = 1.0
a2 = 2.0
w1 = 1.0
w2 = 1.0
t = linspace(0,17,num=300)
x1 = a1 * sin(w1 * t)
x2 = a2 * cos(w2 * t)
plt.plot(t,x1)
plt.plot(t,x2)
```

Далее изобразим кривую, которая является суммой двух гармоник:

```python
from numpy import *
import matplotlib.pyplot as plt

a1 = 1.0
a2 = 1.0
w1 = 1.0
w2 = 1.0
t = linspace(0,17,num=300)
x1 = a1 * cos(w1 * t)
x2 = a2 * cos(w2 * t)
y = x1 + x2
plt.plot(t,y)
plt.plot(t,x1)
plt.plot(t,x2)
```

Пока гармоники совпадают по фазе - график суммарной функции просто удваивает амплитуду. Что произойдет, если фазы гармоник разойдутся на 180 градусов?

```python
from numpy import *
import matplotlib.pyplot as plt

a1 = 1.0
a2 = 1.0
w1 = 1.0
w2 = 1.0
ph1 = 3.1415196
ph2 = 0
t = linspace(0,17,num=300)
x1 = a1 * cos(w1 * t + ph1)
x2 = a2 * cos(w2 * t + ph2)
y = x1 + x2
plt.plot(t,y)
plt.plot(t,x1)
plt.plot(t,x2)
```

Что произойдет, если фазы гармоник разойдутся на 90 градусов?

```python
from numpy import *
import matplotlib.pyplot as plt

a1 = 1.0
a2 = 1.0
w1 = 1.0
w2 = 1.0
ph1 = 3.1415196 / 2
ph2 = 0
t = linspace(0,17,num=300)
x1 = a1 * cos(w1 * t + ph1)
x2 = a2 * cos(w2 * t + ph2)
y = x1 + x2
plt.plot(t,y)
plt.plot(t,x1)
plt.plot(t,x2)
```

Вернем на место фазы, при этом поменяем отношение частот слагаемых гармоник. Также для наглядности увеличим масштаб времени:

```python
from numpy import *
import matplotlib.pyplot as plt

a1 = 1.0
a2 = 1.0
w1 = 1.0
w2 = 1.2
ph1 = 0
ph2 = 0
t = linspace(0,90,num=300) # масштаб времени был увеличен до 60 единиц
x1 = a1 * cos(w1 * t + ph1)
x2 = a2 * cos(w2 * t + ph2)
y = x1 + x2
plt.plot(t,y)
plt.plot(t,x1)
plt.plot(t,x2)
```
