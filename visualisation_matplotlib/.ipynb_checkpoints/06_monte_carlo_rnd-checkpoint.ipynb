# Моделирование случайных процессов

Нахождение числа $\pi$ методом Монте-Карло

```python
x = [] ; y = []

N = 100

for i in range(N):
    x.append(random.random()); y.append(random.random())

plot(x,y,'ro')
```

```python
xr = [] ; yr = []
Nk = 0

for i in range(N):
    if x[i] ** 2 + y[i] ** 2 <=1:
        Nk = Nk + 1
        xr.append(x[i]); yr.append(y[i])

4.0 * Nk / N # искомое пи

plot(x,y,'ob',xr,yr,'or')
```
