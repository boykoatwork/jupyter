## Первый график: интерфейс Pylab

После активации pylab, оболочка готова к отрисовке.  Включение режима ``%pylab`` выполняет несколько действий, в том числе импорт библиотеки ``pylab`` в текущее пространство имен. Для ясности в данной работе это проделывается непосредственно. Так же следуюет выполнить импорт библиотеки ``numpy`` для манипуляций с массивами, на основе которых будут строится графики:

**специальное примечание:** номера ячеек ввода и вывода совпадать не будут - это сделано намеренно.

```python
import pylab
import numpy as np
```

Построим простейшую синусоиду:

```python
x = np.linspace(0, 20, 1000)  # задаем массив из 1000 равномерно распределенных величин от 0 до 20
y = np.sin(x)
```

```python
pylab.plot(x, y)
pylab.show()
```

### Настройка графика: пределы осей

Немного улучшим вид графика - для начала поменяем пределы осей с помощью ``xlim()`` и ``ylim()``

```python
pylab.plot(x, y)
pylab.xlim(5, 15)
pylab.ylim(-0.5, 0.9)
pylab.show()
```

### Изменения названий осей и подписи

Изменить названия осей и задать заголовок можно следующим образом:

```python
pylab.plot(x, y)

pylab.xlabel('this is x!')
pylab.ylabel('this is y!')
pylab.title('My First Plot')
pylab.show()
```

В качестве надписей также можно использовать символы LaTeX (при условии, что он установлен на компьютере):

```python
y = np.sin(2 * np.pi * x)
pylab.plot(x, y)
pylab.title(r'$\sin(2 \pi x)$')  #  `r` перед текстовой переменной указывает на то, что это raw string - любые спецсимволы толкуются именно как символы
pylab.show()
```

### Стили кривых

Можно изменять настройки стилей кривых в широком диапазоне. Например можно изменять их цвета:

```python
pylab.plot(x, y, '*g', label="это график")  # сплошная красная (r - red) линия
pylab.xlim(0, 10)
pylab.ylim(-1.2, 1.2)

pylab.xlabel('this is x!')
pylab.ylabel('this is y!')
pylab.legend(loc="lower left")
pylab.title('My First Plot')
```

В том числе доступны следующие цвета:

     'r' = красный
     'g' = зеленый
     'b' = синий
     'c' = голубой (cyan)
     'm' = пурпурный (magenta)
     'y' = желтый
     'k' = черный
     'w' = белый

Стили линий могут быть следующими:

     '-' = сплошной
     '--' = пунктир
     ':' = точечный
     '-.' = точка-тире
     '.' = точечный 2
     'o' = заполненными кружками
     '^' = заполненными треугольниками

и так далее.

Больше информации можно получить задав в оболочке магическую команду ``pylab.plot?`` 

```python
pylab.plot?
```

### Задание легенды чертежа

На одном и том же графике может быть показано несколько линий. В таком случае можно использовать легенду, чтобы обозначить что означает каждая из линий:

```python
x = np.linspace(0, 20, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

pylab.plot(x, y1, '-b', label='sine')
pylab.plot(x, y2, '-r', label='cosine')
pylab.legend(loc='upper right')
pylab.ylim(-1.5, 2.0)
pylab.show()
```
