https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html

```python
edible = []
non_edible = []

for i in range(100):
    m = 60 + np.random.normal() * 9
    r = 55 + np.random.normal() * 10
    edible.append((m,r))
```

```python
x = [x[0] for x in edible]
y = [x[1] for x in edible]

plt.plot(x,y,'or')
plt.show()
```

```python
non_edible = []

for i in range(100):
    m = abs(20 + np.random.normal() * 10)
    r = abs(20 + np.random.normal() * 10)
    non_edible.append((m,r))
```

```python
xn,yn = [x[0] for x in non_edible], [x[1] for x in non_edible]

plt.plot(xn,yn,'ob')
plt.show()
```

### 2. Apples and sklearn

```python
from sklearn.neighbors import KNeighborsClassifier
```

Creating model:

```python
model = KNeighborsClassifier(n_neighbors=3)
```

Set model on array:

```python
edible[:3], non_edible[:3]
```

```python
apples = []
labels = []
for i in edible:
    apples.append(i)
    labels.append(1)

for j in non_edible:
    apples.append(j)
    labels.append(0)
    
apples[98:102],labels[98:102]
```

```python
# Train the model using the training sets
model.fit(apples,labels)
```

### Classification

```python
#Predict Output
predicted= model.predict([[x_new,y_new]])
print(predicted)
```
